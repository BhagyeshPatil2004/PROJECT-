```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo11n.pt")

# Train the model
train_results = model.train(
    data=r"C:\ImgRegYolo11\Tumor Detection\data.yaml",  # path to dataset YAML
    epochs=20,  # number of training epochs
    imgsz=640,  # training image size
    device=0,  # device to run on, i.e. device=0 or device=0,1,2,3 or device=cpu
)
```

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2

# Load the model
model = YOLO("C:\\ImgRegYolo11\\runs\\detect\\train\\weights\\best.pt")

# Perform object detection on an image
results = model(r"C:\ImgRegYolo11\DemoImg\image4.jpg")

# Convert the results image to a format that matplotlib can handle (BGR to RGB)
img = results[0].plot()  # This will generate an image with boxes drawn

# Display the image in Jupyter Notebook
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for matplotlib
plt.axis('off')  # Hide axes
plt.show()
```
